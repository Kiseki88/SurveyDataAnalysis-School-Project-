```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('surveydata.csv')
```

Here i have removed a Timestamp Column since it is inrelevent in my assignment project, and make sure it is pass through to other derivative with inplace = True.

```python
qty_df = df.iloc[:, :10]
qual_df = df.iloc[:, 10:]
```

```python
qty_df.info()
```

```python
# Create a dictionary mapping old column names to new column names
columns_mapping = {
    "How often do you use Google Drive for storing files?": "Q1",
    "How has your choice of cloud storage model influenced your satisfaction rate with Google Drive?": "Q2",
    "On a scale of 1 to 5, how satisfied are you with the current cloud storage model you're using with Google Drive?": "Q3",
    "How important is data security to you when choosing a cloud storage model for Google Drive?": "Q4",
    "How easy is it to access your files across different devices with the current cloud storage model you're using with Google Drive?": "Q5",
    "Have you experienced differences in satisfaction rate with Google Drive based on the chosen cloud storage model?": "Q6",
    "Which aspect of the selected cloud storage model do you think has the most significant impact on satisfaction rate with Google Drive?": "Q7",
    "To what extent do you believe the current cloud storage model affects the overall user experience of Google Drive?": "Q8",
    "How satisfied are you with the integration of third-party apps and services with the current cloud storage model for Google Drive?": "Q9",
    "Which cloud storage model do you think offers the best compatibility and optimization for use with Google Drive?": "Q10"
}

# Rename the columns in the DataFrame
qty_df.rename(columns=columns_mapping, inplace=True)
```

```python
qty_df
```

```python
qty_df['Q1'].replace(['Rarely', 'Monthly', 'Weekly', 'Daily'], [1, 2, 3, 4], inplace=True)

qty_df['Q2'].replace(['Negatively', 'No noticeable impact', 'Positively'], [1, 2, 3], inplace=True)

qty_df['Q3'].replace(['Very Dissatisfied', 'Dissatisfied', 'Neutral', 'Satisfied', 'Very Satisfied'], [1, 2, 3, 4, 5], inplace=True)

qty_df['Q4'].replace(['Not important', 'Somewhat important', 'Important', 'Very important', 'Extremely important'], [1, 2, 3, 4, 5], inplace=True)

qty_df['Q5'].replace(['Very Difficult', 'Difficult', 'Neutral', 'Easy', 'Very Easy'], [1, 2, 3, 4, 5], inplace=True)

qty_df['Q6'].replace(['No, not at all', 'Not sure', 'Yes, somewhat', 'Yes, significantly'], [1, 2, 3, 4], inplace=True)

qty_df['Q7'].replace(['Upload/download speed', 'Collaboration tools', 'File organization features', 'Reliability and uptime'], [1, 2, 3, 4], inplace=True)

qty_df['Q8'].replace(['Not At All', 'Slightly', 'Moderately', 'Significantly', 'Extremely'], [1, 2, 3, 4, 5], inplace=True)

qty_df['Q9'].replace(['Very Dissatisfied', 'Dissatisfied', 'Neutral', 'Satisfied', 'Very Satisfied'], [1, 2, 3, 4, 5], inplace=True)

qty_df['Q10'].replace(['iCloud', 'Dropbox'], [1, 2], inplace=True)
```

Since my survey has two parts invloved , quantitative and qualitative. I have slice the dataFrame into two parts one for qty(quantitative) and other for quali(qualitative)

**df.iloc[:, :10] selects rows from all rows (: means select all rows) and columns from index 0 to 10 (inclusive). Then we pass the data back to our qty_df variable.**

```python
qty_df.describe()
```

```python
qty_df = qty_df.astype(str)

# Determine the number of rows for subplots
num_rows = len(qty_df.columns)

# Create subplots with the specified layout and margins
fig, axes = plt.subplots(num_rows, 1, figsize=(10, num_rows * 4), sharex=False)

# Iterate over each column and plot as a bar plot
for idx, column in enumerate(qty_df.columns):
    ax = axes[idx]

    ax.bar(qty_df.index, qty_df[column])
    ax.set_xlabel("Respondents")
    ax.set_title(f"Survey Data - {column}")

    # Set margins around the plot
    ax.margins(x=0.1, y=0.3)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**Using a loop, I've generated a plot for each column. Now, to clarify, let's create a plot for each column and then provide an explanation**
